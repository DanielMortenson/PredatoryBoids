```python
from functions import hunt
import matplotlib.pyplot as plt
```

## Change these motivation arrays to alter the behavior of the predators and the prey

```python

#separation, fleeing, conformity, cohesion
b_motivation = [2.5, 3, 2, 1]
#separation, hunting, conformity, cohesion
p_motivation = [2, 3, 1,1]
```

## Run the animation with hunt and additional parameters

```python
n_captured = hunt(  n_boids = 500, 
                n_preds = 20, 
                time_steps = 150,
                pred_swarm  = True,
                circle_start = True,
                anim = True,
                sight_boids = 20,
                sight_pred = 50,
                x_lim = [-200,200],
                y_lim = [-150,150],
                b_motivation = b_motivation,
                p_motivation = p_motivation,
                anim_type = ".gif") #anim type can also be "".mp4"
print(f"Total prey captured: {n_captured[-1]}")
```

## Plot the number of prey captured over time

```python
fig,ax = plt.subplots(1,figsize = (12,8))
plt.plot(n_captured)
plt.title("Number of Prey Captured",size = 25)
plt.xlabel("Time Steps")
plt.ylabel("Prey")
```
